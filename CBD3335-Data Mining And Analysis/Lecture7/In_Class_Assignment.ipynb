```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch 
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_csv('train.csv')[['SalePrice','LotArea']]
```

```python
df.head(2)
```

```python
plt.scatter(df['LotArea'], df['SalePrice'])
```

```python
X = df[['LotArea']]
y = df[['SalePrice']]
```

```python
z= sch.linkage(X, 'ward')
```

```python
dendogram = sch.dendrogram(z, truncate_mode='lastp', p=12, leaf_rotation=45., leaf_font_size=15., show_contracted=True)
```

```python
hc = AgglomerativeClustering(n_clusters=3)
```

```python
y_hc = hc.fit(df)
print(y_hc)
```

```python
labels = y_hc.labels_
print(labels)
```

```python
plt.scatter(df['LotArea'], df['SalePrice'], c=labels, cmap='viridis')
```
