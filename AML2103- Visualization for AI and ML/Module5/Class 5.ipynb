### 1. Line Graph:

Line charts are used to display quantitative values over a continuous time period and show information as a series.
The value is placed on the y-axis, while the x-axis is the timescale.
For smaller time periods, vertical bar charts might be the better choice.

#### 1.6) Legends and Text:

```python
plt.figure()
plt.plot([1,2,3],'m*-',label = "USA") # USA
plt.plot([6,7,4],'gs:',label = "Germany") # Germany

plt.title('This is the title of the figure', fontweight = 'bold', fontsize = 12)
plt.xlabel('This is the x-axis ')
plt.ylabel('This is the y-axis ', rotation = 0, labelpad = 45)

plt.legend()
plt.show()
```

### 1.7) Stockprices data:

Visualize the trends of each company's closing stock price, over time

```python
df = pd.read_csv(r'./stockprices.csv') # , index_col=0
```

```python
df
```

```python
df_graph = df[["Name", "close", "date"]]
```

```python
plt.figure(figsize=(20,20))
for x in df_graph["Name"].unique():
    plt.plot('date','close', data = df_graph[df_graph["Name"] == x],linewidth = 2, markersize = 4,label= x)            

plt.title('Stock trends of 5 companies', fontweight='bold')
plt.ylabel("Closing price in $")
plt.yticks(np.arange(0, df["close"].max(), 70))
plt.xticks(np.arange(0, len(df["date"].unique()), 60), rotation = 45)
plt.grid()
plt.legend(fontsize=20, loc="upper center")
plt.show()
```

```python
plt.figure(figsize=(20,20))

for item in df["Name"].unique():
    name = item
    if item == 'AMZN':
        name = 'Amazon'
    if item == 'AAPL':
        name = 'Apple'
    if item == 'GOOGL':
        name = 'Google'
    if item == 'FB':
        name = 'Facebook'
    if item == 'MSFT':
        name = 'Microsoft'
    plt.plot('date', 'close', data=df[df["Name"]==item], label=name)
    
plt.title('Stock trends of 5 companies', fontweight='bold')
plt.ylabel("Closing price in $")
plt.yticks(np.arange(0, df["close"].max(), 70))
plt.xticks(np.arange(0, len(df["date"].unique()), 60), rotation = 45)
plt.grid()
plt.legend(fontsize=20, loc="upper center")
plt.show()
```
