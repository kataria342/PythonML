## Line Graph

```python
df.sort_values('id', inplace=True)
```

```python
# txn_id v/s txn_total
from matplotlib import pyplot as plt

plt.plot('id', 'txn_total', 'm*--', data=df)
plt.xlabel('Transaction ID')
plt.ylabel('Transaction Total')
plt.title('Line Graph of Transaction in $')
plt.legend()
plt.show()
```

```python
# txn_id v/s txn_total
# Function to annotate every marker:
from matplotlib import pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot('id', 'txn_total', 'm*--', data=df, label='Transaction Amounts')
plt.xlabel('Transaction ID')
plt.ylabel('Transaction Total')
plt.title('Line Graph of Transaction in $')
plt.legend()
# plt.show()


for x, y in zip(df['id'], df['txn_total']):
    label = "{:.2f}".format(y)
    plt.annotate(label, (x, y))
```

Grouped bar graph/ multiple-category bar graph

```python
import numpy as np
course_list = ['CSE', 'ME', 'ECE']
boysCount = [90, 150, 110]
girlsCount = [100, 50, 90]

np.arange(len(course_list))
w = 0.4

bar1 = np.arange(len(course_list))    # 0,1,2
bar2 = [(i+w) for i in bar1] 


plt.bar(bar1, boysCount, w, label= 'Boys')
#plt.bar(bar1+w, girlsCount, w)
plt.bar(bar2, girlsCount, w, label='Girls')
plt.title('#Students Vs Courses')
plt.xlabel('Courses')
plt.ylabel('#Studetns')
plt.legend()
plt.xticks(bar1 + (w/2), course_list)
```
