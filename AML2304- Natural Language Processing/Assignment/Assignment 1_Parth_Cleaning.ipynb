```python
df = df[['content']].dropna()
df.head()
```

### N-grams

```python
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
import collections
```

```python
trigrams = ngrams(word_tokenize(df['content'].sum()), 3)
trigrams_freq = collections.Counter(trigrams)
trigrams_freq.most_common()
```

### Number Removal

```python
import re
def removeNumber(myString):
    rem_num = re.sub(r'\d+', '', myString)
    return rem_num

df['content']=df['content'].apply(lambda x: removeNumber(x))
df.head()
```

### Remove Punctuation

```python
import string
punctuation = string.punctuation
```

```python
def remove_punctuation(text):
    no_punct=[words for words in text if words not in punctuation]
    words_wo_punct=''.join(no_punct)
    return words_wo_punct
df['content'] = df['content'].apply(lambda x: remove_punctuation(x))
df.head()
```

### Tokenization

```python
import re
def tokenize(text):
    split=re.split("\W+",text) 
    return split
df['content']=df['content'].apply(lambda x: tokenize(x.lower()))
df.head()
```

### Stop words

```python
import nltk
stopword = nltk.corpus.stopwords.words('english')
print(stopword[:11])
```

```python
def remove_stopwords(text):
    text=[word for word in text if word not in stopword]
    return text
df['content'] = df['content'].apply(lambda x: remove_stopwords(x))
df.head()
```

### Lemmatize/ Stemming

```python
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
words = ["write", "writee", "writing"]
stemmer = PorterStemmer()
for word in words:
    print(word + " = " + stemmer.stem(word))
```

### checking popular words

```python
from collections import Counter

def popular_words(series):
    df['content'] = series.apply(lambda x:str(x).split())
    top = Counter([item for sublist in df['content'] for item in sublist])
    temp = pd.DataFrame(top.most_common(10))
    temp.columns = ['Common_words','count']
    return temp

print(popular_words(df['content']))
```
