#### ***Finding maximum and minimum in matrix***

```python
mat1 = np.array([[1,2,3],[2,4,6],[7,8,9]])
mat1
```

```python
print('Minimum value in matrix :',np.min(mat1))
print('Maxium value in matrix :',np.max(mat1))
print('Minimum value in rows of the matrix :',np.min(mat1, axis=0))
print('Minimum value in column of the matrix :',np.min(mat1, axis=1))
print('Maximum value of 1st column of the matrix :',np.max(mat1, axis=1)[0])
print('Maximum value of 1st column of the matrix :',np.max(mat1[0]))
print('Mean of the matrix :', np.mean(mat1))
print('Mean of the first row in matrix :', np.mean(mat1, axis=0))
print('Variance of the matrix :', np.var(mat1))
print('Standard deviation of matrix : ', np.std(mat1))
```

##### Generating random values

```python
import matplotlib.pyplot as plt
randList = np.random.randint(0, 1100, 30) #start=5, end=11(excluded), Count=3
print(type(randList))
plt.hist(randList, bins=10)
plt.show()
#print(randList)
```

```python
normal = np.random.normal(0.0,1.0,3000)
```

```python
plt.hist(normal)
```

```python
logistic = np.random.logistic(0.0,1.0,3000)
plt.hist(logistic)
plt.show()
```

```python
fig , (ax1, ax2) = plt.subplots(2,1)
ax1.hist(normal)
ax2.hist(logistic)
```

```python
import seaborn as sns

fig, axis = plt.subplots(1, 2, figsize=(15, 5), sharey=True)

# Normal
sns.histplot(ax=axis[0], data=normal, element='poly')
axis[0].set_title('Normal Distribution')

# Logistic
sns.histplot(ax=axis[1],data=logistic, element='poly')
axis[1].set_title('logistic Distribution')
```

```python
plt.figure(figsize=(20,10))
plt.hist(normal,histtype="step", label="normal")
plt.hist(logistic,histtype="step", label="logistic")
plt.legend()
plt.show()
```

```python
uniform = np.random.uniform(1.0, 2.0, 3000)
```

```python
plt.figure(figsize=(20,10))
plt.hist(uniform, histtype="step", label="normal", color='blue')
plt.hist(logistic, histtype="step", label="logistic", color='red')
plt.legend()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6), sharex=True)
sns.histplot(logistic, kde=True, ax=ax1)
ax1.set_ylabel('Logistic')

sns.histplot(uniform, kde=True, ax=ax2)
ax2.set_ylabel('Uniform')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.kdeplot(logistic, fill=True, color="r")
sns.kdeplot(uniform, fill=True, color="b")
plt.show()
```
